```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from google.colab import files
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    confusion_matrix, classification_report,
    roc_auc_score, average_precision_score,
    RocCurveDisplay, PrecisionRecallDisplay
)

np.random.seed(42)
tf.random.set_seed(42)

# =========================
# 1) LOAD DATA (Uploaded in Colab)
# =========================
# If you already uploaded, this will usually exist:
# Changed from .csv.zip to .csv, assuming it's a plain CSV file misnamed as a zip.
csv_path = "/content/creditcard.csv"

# If not found, this will let you upload now (safe even if you already uploaded).
import os
if not os.path.exists(csv_path):
    uploaded = files.upload()  # choose creditcard.csv
    csv_path = list(uploaded.keys())[0]

df = pd.read_csv(csv_path)
print("Data shape:", df.shape)
print(df["Class"].value_counts())

# Handle potential NaN values in the 'Class' column before converting to int
if df['Class'].isnull().any():
    print(f"Warning: {df['Class'].isnull().sum()} NaN values found in 'Class' column. Dropping rows with NaNs.")
    df.dropna(subset=['Class'], inplace=True)

X = df.drop("Class", axis=1)
y = df["Class"].astype(int)

# =========================
# 2) HELPERS
# =========================
def preprocess_fit_transform(X_train, X_other):
    """Fit scaler on train, transform train and other (val/test)."""
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_other_scaled = scaler.transform(X_other)
    return X_train_scaled, X_other_scaled, scaler

def compute_class_weights(y_arr):
    y_arr = np.array(y_arr).astype(int)
    neg, pos = np.bincount(y_arr)
    total = neg + pos
    return {0: total / (2 * neg), 1: total / (2 * pos)}

def build_dnn(input_dim):
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(input_dim,)),
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.4),

        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.3),

        tf.keras.layers.Dense(32, activation="relu"),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.2),

        tf.keras.layers.Dense(16, activation="relu"),
        tf.keras.layers.Dense(1, activation="sigmoid")
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="binary_crossentropy",
        metrics=[
            tf.keras.metrics.AUC(name="roc_auc"),
            tf.keras.metrics.AUC(name="pr_auc", curve="PR")
        ]
    )
    return model

def evaluate_model(model, X_test, y_test, title=""):
    y_pred_prob = model.predict(X_test, verbose=0).ravel()
    y_pred = (y_pred_prob >= 0.5).astype(int)

    print("\n" + "="*70)
    print("EVALUATION", f"({title})" if title else "")
    print("="*70)

    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, digits=4))

    roc_auc = roc_auc_score(y_test, y_pred_prob)
    pr_auc = average_precision_score(y_test, y_pred_prob)

    print(f"ROC-AUC: {roc_auc:.4f}")
    print(f"PR-AUC : {pr_auc:.4f}")

    RocCurveDisplay.from_predictions(y_test, y_pred_prob)
    plt.title(f"ROC Curve {title}".strip())
    plt.show()

    PrecisionRecallDisplay.from_predictions(y_test, y_pred_prob)
    plt.title(f"Precision-Recall Curve {title}".strip())
    plt.show()

    return roc_auc, pr_auc

# =========================
# 3) EXPERIMENT 1: 80/20 split
# =========================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
X_train_s, X_test_s, scaler = preprocess_fit_transform(X_train, X_test)

model = build_dnn(X_train_s.shape[1])
class_weights = compute_class_weights(y_train)

history = model.fit(
    X_train_s, y_train,
    epochs=20,
    batch_size=2048,
    class_weight=class_weights,
    verbose=1
)

plt.plot(history.history["loss"])
plt.title("Training Loss (80/20)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()

evaluate_model(model, X_test_s, y_test, title="80/20")

# =========================
# 4) EXPERIMENT 2: 70/30 split
# =========================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)
X_train_s, X_test_s, scaler = preprocess_fit_transform(X_train, X_test)

model = build_dnn(X_train_s.shape[1])
class_weights = compute_class_weights(y_train)

history = model.fit(
    X_train_s, y_train,
    epochs=20,
    batch_size=2048,
    class_weight=class_weights,
    verbose=1
)

evaluate_model(model, X_test_s, y_test, title="70/30")

# =========================
# 5) EXPERIMENT 3: Train/Val/Test = 60/20/20 with EarlyStopping
# =========================
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.25, stratify=y_temp, random_state=42
)  # 0.25 of 0.8 = 0.2

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_val_s   = scaler.transform(X_val)
X_test_s  = scaler.transform(X_test)

model = build_dnn(X_train_s.shape[1])
class_weights = compute_class_weights(y_train)

early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_pr_auc",
    mode="max",
    patience=5,
    restore_best_weights=True
)

history = model.fit(
    X_train_s, y_train,
    validation_data=(X_val_s, y_val),
    epochs=50,
    batch_size=2048,
    class_weight=class_weights,
    callbacks=[early_stop],
    verbose=1
)

plt.plot(history.history["loss"], label="Train")
plt.plot(history.history["val_loss"], label="Validation")
plt.legend()
plt.title("Loss Curves (60/20/20)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()

evaluate_model(model, X_test_s, y_test, title="60/20/20 + EarlyStop")

# =========================
# 6) EXPERIMENT 4: Stratified 5-Fold CV (AUC avg)
# =========================
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
roc_scores, pr_scores = [], []

X_np = X.values
y_np = y.values.astype(int)

for fold, (train_idx, test_idx) in enumerate(skf.split(X_np, y_np), start=1):
    X_train, X_test = X_np[train_idx], X_np[test_idx]
    y_train, y_test = y_np[train_idx], y_np[test_idx]

    scaler = StandardScaler()
    X_train_s = scaler.fit_transform(X_train)
    X_test_s  = scaler.transform(X_test)

    model = build_dnn(X_train_s.shape[1])
    class_weights = compute_class_weights(y_train)

    model.fit(
        X_train_s, y_train,
        epochs=20,
        batch_size=2048,
        class_weight=class_weights,
        verbose=0
    )

    y_pred_prob = model.predict(X_test_s, verbose=0).ravel()
    roc_scores.append(roc_auc_score(y_test, y_pred_prob))
    pr_scores.append(average_precision_score(y_test, y_pred_prob))

print("\n" + "="*70)
print("5-FOLD CV RESULTS")
print("="*70)
print(f"ROC-AUC: {np.mean(roc_scores):.4f} \u00b1 {np.std(roc_scores):.4f}")
print(f"PR-AUC : {np.mean(pr_scores):.4f} \u00b1 {np.std(pr_scores):.4f}")
```
